```python
from google.colab import drive
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
drive.mount('/content/gdrive')
```

```python
img = cv2.imread('/content/gdrive/My Drive/pj1/영상이미지/Images/lena.jpg')
RGB_Img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gImg = cv2.cvtColor(RGB_Img, cv2.COLOR_BGR2GRAY)
print(gImg.shape)
plt.imshow(gImg)
```

```python
kernel3 = np.ones((3,3), np.float32)/9
kernel5 = np.ones((5,5), np.float32)/25
```

```python
kernel3[0][0] = 1/9
kernel3[0][1] = 1/9
kernel3[0][2] = 1/9
kernel3[1][0] = 1/9
kernel3[1][1] = 1/9
kernel3[1][2] = 1/9
kernel3[2][0] = 1/9
kernel3[2][1] = 1/9
kernel3[2][2] = 1/9
```

```python
outImg3 = cv2.filter2D(gImg, -1, kernel3)
cv2_imshow(outImg3)
```

```python
outImg5 = cv2.filter2D(gImg, -1, kernel5)
cv2_imshow(outImg5)
```

```python
outImg33 = cv2.filter2D(outImg3, -1, kernel3)
cv2_imshow(outImg33)
```
