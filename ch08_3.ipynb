```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('/content/gdrive/My Drive/pj1/영상이미지/Images/rice.bmp')

RGB_Img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gImg = cv2.cvtColor(RGB_Img, cv2.COLOR_BGR2GRAY)
print(gImg.shape)
plt.imshow(gImg, cmap='gray')  # cmap 파라미터를 'gray'로 설정
plt.axis('off')  # 축 숨기기 (선택 사항)
plt.show()
```

```python
kernel = np.ones((3,3), np.uint8)
erodeImg = cv2.erode(gImg, kernel, iterations = 1)
cv2_imshow(erodeImg)
```

```python
gBin = np.zeros((gImg.shape[0], gImg.shape[1]), dtype=np.ubyte)
```

```python
threshold = 120

for h in range(gImg.shape[0]):
  for w in range(gImg.shape[1]):
    if(gImg[h,w] > threshold):
      gBin[h,w] = 255

cv2_imshow(gBin)
```

```python
kernel = np.ones((3,3), np.uint8)
erodeImg = cv2.erode(gBin, kernel, iterations = 1)
cv2_imshow(erodeImg)
```

```python
kernel = np.ones((3,3), np.uint8)
dilateImg = cv2.dilate(erodeImg,kernel, iterations = 1)
cv2_imshow(dilateImg)
```

```python
kernel = np.ones((3,3), np.uint8)
openImg = cv2.morphologyEx(gBin, cv2.MORPH_OPEN, kernel)
cv2_imshow(openImg)
```
