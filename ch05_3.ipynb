```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('/content/gdrive/My Drive/pj1/영상이미지/Images/lena.jpg')
RGB_Img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gImg = cv2.cvtColor(RGB_Img, cv2.COLOR_BGR2GRAY)
print(gImg.shape)
plt.imshow(gImg, cmap='gray')  # cmap 파라미터를 'gray'로 설정
plt.axis('off')  # 축 숨기기 (선택 사항)
plt.show()
```

```python
outImg = cv2.GaussianBlur(gImg, (3,3), 3)
cv2_imshow(outImg)
```

```python
outImg = cv2.GaussianBlur(gImg, (7,7), 3)
cv2_imshow(outImg)
```

```python
outImg = cv2.GaussianBlur(gImg, (15,15), 3)
cv2_imshow(outImg)
```
